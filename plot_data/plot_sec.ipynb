```python
f0=pathWD+'TMB2_17.xlsx'
f1=pathWD+'V12T.xlsx'
f2=pathWD+'V10G.xlsx'
f3=pathWD+'V12G.xlsx'
f4=pathWD+'V10P.xlsx'
f5=pathWD+'V12P.xlsx'

df0_sec = pd.read_excel(f0, header=0, index_col=0)
df1_sec = pd.read_excel(f1, header=0, index_col=0)
df2_sec = pd.read_excel(f2, header=0, index_col=0)
df3_sec = pd.read_excel(f3, header=0, index_col=0)
df4_sec = pd.read_excel(f4, header=0, index_col=0)
df5_sec = pd.read_excel(f5, header=0, index_col=0)
```

```python
df0_sec
```

```python
# copy the data
df0_scaled = df0_sec.copy()
df1_scaled = df1_sec.copy()
df2_scaled = df2_sec.copy()
df3_scaled = df3_sec.copy()
df4_scaled = df4_sec.copy()
df5_scaled = df5_sec.copy()

# apply normalization techniques on Column 1
df0_scaled['mAU'] = df0_scaled['mAU'] /df0_scaled['mAU'].abs().max()
df1_scaled['mAU'] = df1_scaled['mAU'] /df1_scaled['mAU'].abs().max()
df2_scaled['mAU'] = df2_scaled['mAU'] /df2_scaled['mAU'].abs().max()
df3_scaled['mAU'] = df3_scaled['mAU'] /df3_scaled['mAU'].abs().max()
df4_scaled['mAU'] = df4_scaled['mAU'] /df4_scaled['mAU'].abs().max()
df5_scaled['mAU'] = df5_scaled['mAU'] /df5_scaled['mAU'].abs().max()

# slice on index
df0_scaled = df0_scaled.loc[(df0_scaled.index>6.0)&(df0_scaled.index<18.0)]
df1_scaled = df1_scaled.loc[(df1_scaled.index>6.0)&(df1_scaled.index<18.0)]
df2_scaled = df2_scaled.loc[(df2_scaled.index>6.0)&(df2_scaled.index<18.0)]
df3_scaled = df3_scaled.loc[(df3_scaled.index>6.0)&(df3_scaled.index<18.0)]
df4_scaled = df4_scaled.loc[(df4_scaled.index>6.0)&(df4_scaled.index<18.0)]
df5_scaled = df5_scaled.loc[(df5_scaled.index>6.0)&(df5_scaled.index<18.0)]

# view normalized data
display(df0_scaled)
```

```python

#sns.set_style(\"whitegrid\")
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(5,6))

df0_scaled['mAU'].plot(ax=axes[0,0],color='purple',linewidth=1, label='TMB2.17', fontsize='8')
axes[0,0].arrow( 14.182604, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)
df1_scaled['mAU'].plot(ax=axes[0,1],color='silver',linewidth=1, label='V12T', fontsize='8')
axes[0,1].arrow( 14.49928, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)
df2_scaled['mAU'].plot(ax=axes[1,0],color='silver',linewidth=1, label='V10G', fontsize='8')
axes[1,0].arrow( 14.450909, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)
df3_scaled['mAU'].plot(ax=axes[1,1],color='silver',linewidth=1, label='V12G', fontsize='8')
axes[1,1].arrow( 14.466699, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)
df4_scaled['mAU'].plot(ax=axes[2,0],color='silver',linewidth=1, label='V10P', fontsize='8')
axes[2,0].arrow( 14.443605, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)
df5_scaled['mAU'].plot(ax=axes[2,1],color='silver',linewidth=1, label='V12P', fontsize='8')
axes[2,1].arrow( 13.896688, 0.15, 0.0, -0.1, fc="k", ec="k",head_width=0.4, head_length=0.05)

#plt.xlabel('Retention volume (ml)', fontsize=20)
axes[0,1].legend(prop={'size': 8})
axes[0,0].legend(prop={'size': 8})
axes[1,0].legend(prop={'size': 8})
axes[1,1].legend(prop={'size': 8})
axes[2,1].legend(prop={'size': 8})
axes[2,0].legend(prop={'size': 8})
```
